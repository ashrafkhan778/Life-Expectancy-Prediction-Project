# 1. Distribution of Life Expectancy


```python
plt.figure(figsize=(8, 4))

# 1. Distribution of Life Expectancy
sns.histplot(df1['Life_Expectancy'], kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy (years)')
plt.ylabel('Count')
plt.show()
```

# Insight

This chart is a histogram showing the distribution of life expectancy across different values in your dataset. Here are a few insights that can be derived from this visualization:

Central Tendency: The life expectancy data appears to have a unimodal distribution, with the majority of values clustered around the 70-year mark. This indicates that most countries in the dataset have a life expectancy close to 70 years.

Range: The life expectancy values in the dataset range from about 35 years to 90 years, indicating significant variance across different countries.

Skewness: The distribution shows slight left skewness (a long tail on the left), suggesting that while most countries have relatively high life expectancy, there are some countries with much lower life expectancies (below 50 years).

Outliers: There may be a few outliers on the lower end of life expectancy, particularly in the 30–50 year range, where the frequency drops significantly.

# Insights:

Positive Correlation: There seems to be a general positive correlation between life expectancy and GDP. As GDP increases, life expectancy tends to increase as well. This suggests that countries with higher GDPs may have better access to healthcare, education, and other resources that contribute to longer and healthier lives.

Scatter: However, the relationship is not perfectly linear. There is a significant amount of scatter in the data, indicating that other factors besides GDP may also influence life expectancy. These factors could include public health policies, social determinants of health, and cultural practices.

Outliers: There are a few outliers visible in the plot, which are data points that deviate significantly from the general trend. These could represent countries with unique circumstances that affect their life expectancy and GDP.

# 20 Correlation Heatmap Of Numerical Variables

```python
plt.figure(figsize=(15, 12))
correlation_matrix = df1.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

 # Insights

Positive Correlations: The red and orange colors in the heatmap indicate positive correlations between variables. This means that as one variable increases, the other variable also tends to increase. For example, there appears to be a strong positive correlation between "Adult Mortality" and "Polio".

Negative Correlations: The blue and purple colors in the heatmap indicate negative correlations between variables. This means that as one variable increases, the other variable tends to decrease. For example, there appears to be a strong negative correlation between "GDP" and "Under Five Deaths".

No Correlation: The white color in the heatmap indicates no correlation between the variables.
Strength of Correlation: The intensity of the color indicates the strength of the correlation. Darker colors (red, blue, or purple) represent stronger correlations, while lighter colors represent weaker correlations.

Data Limitations: The data used for this heatmap may have limitations, and further analysis would be needed to draw more definitive conclusions.
